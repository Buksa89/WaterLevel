```python
df = pd.read_parquet("data/data.parquet.gzip")
means = df.groupby(by="kilometer").mean()
means
df = df.merge(means, how="left", on="kilometer", suffixes = ['', 'Mean'])
df["deviation"] = df['value'] - df['valueMean']
df.drop(columns=["valueMean","value"], inplace=True)
wisla = df[df["river"] == "Wis≈Ça"]
bug = df[df["river"] == "Bug"]
```

```python
px.line(data_frame=wisla, x='kilometer', y='deviation', animation_frame=wisla.datetime.astype(str),range_y=[-25,25])
```

```python
px.line(data_frame=bug, x='kilometer', y='deviation', animation_frame=bug.datetime.astype(str),range_x=[700,0],range_y=[-35,35])
```
